## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Question: Which pages in the content inventory should a reviewer look at first
# for a refresh, and why?
#
# Unit of analysis: one page (content item) per row.
#
# Decision this improves: which pages a content reviewer with limited capacity
# should prioritize this week.
#
# Action: a human reviews the top-ranked pages and decides whether to refresh,
# expand, or leave as-is — the model does not take action on its own.
#
# Cost of a wrong call: if a declining page is ranked too low, it keeps losing
# visibility unnoticed. If a healthy page is ranked too high, reviewer time is
# wasted on a page that didn't need attention. Since this is a ranked review
# queue (not an automated action), the cost of one wrong ranking is low —
# it costs review time, not traffic.
#
# Why data/ML can help: with thousands of pages and limited reviewer bandwidth,
# a transparent score plus a model-informed ranking helps a human focus on the
# pages most worth reviewing, instead of going through them in arbitrary order.
```
